```python
import random
import re

def validar_letra(letra): ## Esta funcion verifica que lo ingresado sea una letra simple y no una palabra y ademas que sea apha es decir letra 
    if len(letra)==1 and letra.isalpha():
        return True 
    else:
        return False
    
def verificar_letra(palabra,letra): #Esta funcion verifica que la letra ingresada este dentro de la palabra elegida en el ahorcado
    palabra = palabra.upper()
    letra = letra.upper()
    if letra in palabra:
        return True
    else:
        return False 

def mostrador(palabra, lista_mostrar): #Esta funcion me devuelve el valor de la palabra con - en las letras que no adivine y me muestra las que adivine
    palabra_string=palabra #Guardo la palabra original en un string
    palabra=list(palabra) #La transformo en una lista
    palabra=set(palabra) #Luego la transformo en un set
    lista_mostrar=set(lista_mostrar) #transformo las letras ingresadas como lista en un set
    i = palabra & lista_mostrar #tomo su conjuncion se que las letras que se le pasen se encuentran en la palabra y ademas que no se repiten
    i=list(i) #creo una lista con la interseccion
    patron = "[^" + "".join(i) + "]" #creo el patron para luego utilizar una expresion regular para poder crear el string con - en donde no se adivino la letra y que me muestre las que si adivino
    texto_devuelto=re.sub(patron, "-",palabra_string) #ejecuto la expresion regular
    return texto_devuelto #devuelvo el string



def contador_letra(palabra,letra): #cuento la cantidad de veces que se repite la letra en la palabra
    lista_palabra=list(palabra)
    contador=0
    for l in lista_palabra:
        if l==letra:
            contador+=1
    return contador

def muestra_unico(lista): #esta funcion me convierte la lista en un set para eliminar elementos repetidos y me la vuelve a convertir en lista 
    return list(set(lista))
```

```python


    
lista_palabras=["arbol","banana","culo","tigre"] #Creo la listas de palabras a elegir 
palabra_elegida=random.choice(lista_palabras) #utilizo el modulo random para seleccionar una palabra aleatoria
print(f"La palabra elegida es: {palabra_elegida}") #la muestro para corroborar lo que estoy haciendo
print("")
contador=0 #establezco una de las condiciones en las cual el juego finaliza que es cuando el jugador adivina las letras de la palabra haciendo que se incremente el contador
vidas=6 #establezco la segunda condicion correspondiente a las vidas del jugador
lista_letras_elegidas=[] #establezco la lista de letras elegidas en una lista vacia en las cuales almacenare tuplas con el valor de la letra y su respectivo TRUE OR FALSE dependiendo de si la letra se encuentra o no dentro de la palabra
letras_elegidas_mostrador=[] #Establezco la lista donde almacenare las letras que ira ingresando el usuario para luego usarla en la funcion mostrar que me devolvera el texto con la palabra con las letras adivinadas y los _ en las sin adivinar
while contador!=len(palabra_elegida) and vidas!=0: #creo el bucle while con las condiciones de victoria o derrota
    letra=input("Ingrese la letra: ")
    letra=letra.lower()
    if not (validar_letra(letra)): # verifico que ingrese una unica letra y que sea una letra
        print("")
        print("Ingrese una letra valida por favor")
        print("------------------------------------------------------------------------------------")
    elif verificar_letra(palabra_elegida,letra): #verifico que la letra ingresada se encuentre en la palabra 
        letras_elegidas_mostrador.append(letra) #si la letra se encuentra la agrego a la lista que ira a la funcion mostrar
        letras_elegidas_mostrador=muestra_unico(letras_elegidas_mostrador) # verifico que me queden solo los valores unicos
        lista_letras_elegidas.append((letra,True)) #creo la lista de tuplas con la que voy a verificar que ingreso una letra que esta en la palabra y por unica vez 
        lista_a_ordenar=muestra_unico(lista_letras_elegidas) #uso la funcion para las tuplas tmb
        lista_a_ordenar.sort(key=lambda tup: tup[0]) #las ordeno alfabeticamente para que sea mejor para el usuario
        mostrar=mostrador(palabra_elegida,letras_elegidas_mostrador) #guardo el texto en una variable para mostrarle el progreso a el jugador
        if lista_letras_elegidas.count((letra,True))==1: #si la letra elegida tiene contador 1 hace que sume sino quiere decir que ya eligio esa letra entonces no deberia sumar 
            contador+=contador_letra(palabra_elegida,letra)
        print("")
        print(f"Elegista una letra que se encuentra en la palabra, estado: {mostrar}")
        print("")
        print(f"Al momento elegiste las letras: {lista_a_ordenar}") #le muestra el progreso
        print("------------------------------------------------------------------------------------")
    elif not verificar_letra(palabra_elegida,letra): #condicion de restar vidas que la letra no este en la palabra
        lista_letras_elegidas.append((letra,False)) #agrego la letra elegida a la lista
        lista_a_ordenar=muestra_unico(lista_letras_elegidas) #uso la funcion para las tuplas tmb
        lista_a_ordenar.sort(key=lambda tup: tup[0]) #las ordeno alfabeticamente para que sea mejor para el usuario
        vidas-=1
        print("")
        print(f"La letra elegida no se encuentra en su palabra le quedan {vidas} vidas")
        print("")
        print(f"Al momento elegiste las letras: {lista_a_ordenar}") 
        print("------------------------------------------------------------------------------------")
    lista_letras_elegidas.sort(key=lambda tup: tup[0])
    letras_elegidas_mostrador.sort()


print("------------------------------------------------------------------------------------")
if contador==len(palabra_elegida):
    print("Has Ganado!")
else:
    print("Has Perdido! ")
    print("")
    print(f"La palabra era {palabra_elegida}")
print("")

        
print(f"Las letras que has elegido fueron: {lista_letras_elegidas}")
print("")
print(f"Las palabras en letras_elegidas_mostrador fueron: {letras_elegidas_mostrador}")

```
